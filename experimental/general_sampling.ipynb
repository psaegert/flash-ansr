```python
from flash_ansr import SkeletonPool, get_path
import matplotlib.pyplot as plt
```

```python
pool = SkeletonPool.from_config(get_path("configs/v18.0-m/skeleton_pool_val.yaml"))
```

```python
data = pool.sample(n_support=1024)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i in range(3):
    axes[i].hist(data['x_support'][:, i], bins=50, color='blue', alpha=0.7)
    axes[i].set_title(f'X Support Dimension {i+1} Distribution')
```
