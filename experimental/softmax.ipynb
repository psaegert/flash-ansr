```python
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = torch.tensor([0, 0.23, 0.61, 0.89, 1.07, 1.2, 1.3, 1.4, 1.47, 1.56, 1.63])
temperature = 0.3
```

```python
y = torch.softmax(x / temperature, dim=0)
y
```

```python
plt.scatter(x.numpy(), y.numpy(), color='blue', label='Softmax')
```
