```python
from simplipy import SimpliPyEngine
import numpy as np
```

```python
engine = SimpliPyEngine.load('dev_7-3')
```

```python
engine.simplification_rules[:10]
```

```python
# 20 random rules:
random_indices = np.random.choice(len(engine.simplification_rules), size=20, replace=False)
for idx in random_indices:
    print(engine.simplification_rules[idx])
```

```python
# Find out the rule that is     [
#         [
#             "exp",
#             "float(\"-inf\")"
#         ],
#         [
#             "0"
#         ]
#     ],

for i, rule in enumerate(engine.simplification_rules):
    if rule == (
        (
            "exp",
            'float("-inf")'
        ),
        (
            "0",
        )
    ):
        print(f"Found the rule at index {i}!")
```
