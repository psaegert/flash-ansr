```python
import torch
from flash_ansr.train.train import ContrastiveLoss, AILWQ_ContrastiveLoss

from sklearn.datasets import make_classification, make_moons
import matplotlib.pyplot as plt
```

```python
AILWQ_loss = AILWQ_ContrastiveLoss()
loss = ContrastiveLoss()
```

```python
fig, ax = plt.subplots(3, 6, figsize=(20, 10))

for i, ax in enumerate(ax.flatten()):
    X, y = make_classification(n_samples=30, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1)
    AILWQ_loss_value = AILWQ_loss(torch.tensor(X), torch.tensor(y))
    loss_value = loss(torch.tensor(X), torch.tensor(y))
    ax.scatter(X[:, 0], X[:, 1], c=['tab:red' if i == 1 else 'tab:blue' for i in y], s=10)
    ax.set_title(f"AILWQ: {AILWQ_loss_value.item():.2f}\nContrastive: {loss_value.item():.2f}")

fig.tight_layout()
```
