```python
import torch
```

```python
from flash_ansr import SetTransformer
```

```python
set_trf = SetTransformer(
    input_dim=512,
    output_dim=512,
    model_dim=256,
    n_heads=8,
    n_isab=2,
    n_sab=1,
    n_inducing_points=32,
    n_seeds=17,
    ffn_hidden_dim=None,
    dropout=0,
    attn_norm="none",
    ffn_norm="none",
    output_norm="none",
    use_checkpointing=False
)
```

```python
X = torch.randn(3, 13, 512)
example_attn_mask_pos = torch.ones(3, 10).bool()
example_attn_mask_neg = torch.zeros(3, 3).bool()
example_attn_mask = torch.cat([example_attn_mask_pos, example_attn_mask_neg], dim=1)
```

```python
y = set_trf(X)
y.shape
```

```python
y_masked = set_trf(X, attn_mask=example_attn_mask)
y_masked.shape
```

```python
(y == y_masked).any()
```

```python
X_altered = X.clone()
X_altered[:, -3:, :] = torch.randn(3, 3, 512)  # Alter the last 3 elements which are masked out
```

```python
y_altered = set_trf(X_altered)
y_altered.shape
```

```python
y_altered_masked = set_trf(X_altered, attn_mask=example_attn_mask)
y_altered_masked.shape
```

```python
(y_masked == y_altered_masked).all()
```
