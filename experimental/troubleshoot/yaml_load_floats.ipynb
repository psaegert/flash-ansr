```python
import yaml

# Assume your config is in a file named 'config.yaml'
# with open('config.yaml', 'r') as f:
#     config = yaml.safe_load(f)

# For demonstration, using a string again:
yaml_string = """
optimizer:
  kwargs:
    eps: 1e-6
    max_lr: 1e-4
"""

config = yaml.safe_load(yaml_string)

eps_val = config['optimizer']['kwargs']['eps']
lr_val = config['optimizer']['kwargs']['max_lr']

print(f"Parsed eps: {eps_val}, Type: {type(eps_val)}")
print(f"Parsed max_lr: {lr_val}, Type: {type(lr_val)}")
```
