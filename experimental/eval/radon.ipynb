```python
import pandas as pd
import os
import json
from flash_ansr import get_path
```

```python
commands = ['cc', 'hal', 'mi', 'raw']
```

```python
outputs = {}

for c in commands:
    with open(get_path('results', 'radon', f'{c}.json')) as f:
        outputs[c] = json.load(f)
```

```python
pd.DataFrame({k: v['total'] for k, v in outputs['hal'].items()}).T.sort_values('difficulty', ascending=False).round(1)
```

```python
mi_df = pd.DataFrame(outputs['mi']).T.sort_values('mi', ascending=False)
mi_df['mi'] = mi_df['mi'].astype(float).round(1)
mi_df = mi_df[~mi_df.index.str.contains('__init__')]
mi_df
```

```python
raw_df = pd.DataFrame(outputs['raw']).T.sort_values('lloc', ascending=False)
raw_df['comment_raio'] = (raw_df['sloc'].astype(float) / (1 + raw_df['single_comments'].astype(float))).round(1)
raw_df
```
