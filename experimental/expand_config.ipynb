```python
from flash_ansr.utils import load_config, get_path, apply_on_nested
```

```python
config = load_config(get_path("configs/v18.0-m/train.yaml"))
```

```python
def unfold_config(config, max_depth=3):
    def try_load_config(x):
        if isinstance(x, str) and x.endswith(".yaml"):
            return load_config(get_path(x))
        return x

    for _ in range(max_depth):
        config = apply_on_nested(config, try_load_config)
    return config
```

```python
unfold_config(config)
```
