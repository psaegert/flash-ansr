```python
import matplotlib.pyplot as plt
import numpy as np

from flash_ansr import get_path
```

```python
x = np.linspace(-10, 10, 1000)
y = np.exp(-(x**2) / 10) * x
```

```python
fig, ax = plt.subplots(figsize=(2.7, 2), dpi=200)
ax.plot(x, y, color='#555', linewidth=2)

ax.axvspan(-10, -7, color='#e5e5e5', zorder=0)
ax.axvline(-7, color='#ddd', zorder=1)
ax.axvspan(8, 10, color='#e5e5e5', zorder=0)
ax.axvline(8, color='#ddd', zorder=1)

x_points = [-3, 0.5, 2.5, 7]
y_points = [np.exp(-(xp**2) / 10) * xp for xp in x_points]

ax.plot(x_points, y_points, color='tab:red', marker='o', linestyle='None', markersize=6)
ax.plot(x_points, np.full_like(x_points, -1.5), color='tab:blue', marker='o', linestyle='None', markersize=6, clip_on=False, zorder=5)
for xp, yp in zip(x_points, y_points):
    ax.plot([xp, xp], [-1.5, yp], color='#ccc', linestyle='--', linewidth=2, zorder=1)
    


ax.set_xticks([])
ax.set_yticks([])
ax.set_xlim([-10, 10])
ax.set_ylim([-1.5, 1.5])
# ax.set_title('[*, exp, neg, /, pow2, x, 10, x]', pad=10)
# ax.set_xlabel('x', labelpad=10)
# ax.set_ylabel('y', rotation=0, labelpad=15)

plt.savefig(get_path('results', 'visual_abstract', create=True, filename=f'dummy_plot.svg'), bbox_inches='tight', dpi=200)
```
